# Word Frequency Count
Program to count word frequency of a text file in python

```python
# Program to count Word Frequency
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
# Read text file
with open("file.txt") as f:
    text = f.read()
print("\nOriginal Text\n",text)
```

```python
fdist = FreqDist(word_tokenize(text))
print("Output after Frequency Distribution", fdist)
```

```python
fdist
```

```python
print("\n5 Most common Words and their Frequency")
print(fdist.most_common(5))
```

```python
print("\n\n Frequency Distribution Graph with 5 most common words ")
fdist.plot(5)
```

```python
print("\n\n Frequency Distribution Graph with 20 most common words ")
fdist.plot(20)
```

```python
print("\n\n Frequency Distribution Graph with all words ")
fdist.plot()
```
